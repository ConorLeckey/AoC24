# Parse Input

```python

with open("day4_input.txt") as file: 
    lines = list(file)
lines
```

```python
# Part 1

import itertools

def is_xmas_present(coords):
    for c, coord in coords.items():
        expected_x, expected_y = coord
        if lines[expected_y][expected_x] != c: 
            return False
    return True


x_locations = list(itertools.chain.from_iterable([[(x, y) for x, c in enumerate(line) if c == 'X'] for y, line in enumerate(lines)]))

max_x = len(lines[0]) - 1
max_y = len(lines) - 1
xmas_count = 0
for x, y in x_locations:
    x_directions = [0]
    if x >= 3: x_directions.append(-1)
    if max_x - x >= 3: x_directions.append(1)

    y_directions = [0]
    if y >= 3: y_directions.append(-1)
    if max_y - y >= 3: y_directions.append(1)

    for x_dir in x_directions:
        for y_dir in y_directions:
            coords = {c: (x + (i + 1) * x_dir, y + (i + 1) * y_dir) for i, c in enumerate('MAS')}
            if is_xmas_present(coords): xmas_count += 1
xmas_count
```

```python
# Part 2

import itertools

a_locations = list(itertools.chain.from_iterable([[(x, y) for x, c in enumerate(line) if c == 'A'] for y, line in enumerate(lines)]))

max_x = len(lines[0]) - 1
max_y = len(lines) - 1


def is_xmas_present(x, y):
    down_right = lines[y + 1][x + 1]
    down_left = lines[y + 1][x - 1]
    up_right = lines[y - 1][x + 1]
    up_left = lines[y - 1][x - 1]

    return {down_right, up_left} == {'M', 'S'} and {down_left, up_right} == {'M', 'S'}

xmas_count = 0
for x, y in a_locations:
    if x >= 1 and y >=1 and max_x - x >= 1 and max_y - y >= 1:
        if is_xmas_present(x, y): xmas_count += 1
xmas_count
```
